```python
import pandas as pd
```

```python
tr = pd.read_excel('C:/1900-2019.xlsx',skiprows=6,index_col=None)
print(type(tr))
```

```python
#Select useful rows
tr1=tr[['Year','Country','Continent','Dis Mag Value','Dis Mag Scale']]
tr1
```

```python
#get national frequency each year and recreate dataframe
groups=tr1.groupby(['Year','Country'])

c=groups['Country'].count()

tr2=pd.DataFrame({'Frequency':c})
tr2
```

```python
#save as excel file
tr2.to_excel('C:/1900-2019 nation frequency.xlsx')
```

```python
#get total frequency each year and recreate dataframe
groups=tr1.groupby(['Year'])

c=groups['Year'].count()

tr3=pd.DataFrame({'Frequency':c})
tr3
```

```python
#save as excel file
tr3.to_excel('C:/1900-2019 year frequency.xlsx')
```

```python
##get total national frequency and recreate dataframe
groups=tr1.groupby(['Country'])

c=groups['Country'].count()

tr4=pd.DataFrame({'Frequency':c})
tr4.sort_values(by='Frequency',inplace=True,ascending=False)
tr4
```

```python
#select the top 20 countries and save as excel file
tr5=tr4.head(20)
tr5.to_excel('C:/1900-2019 country frequency.xlsx')
```
